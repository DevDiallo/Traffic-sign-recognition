```python
dataset_path = "/content/drive/MyDrive/archive" # Le chemin du dataset contenant sur le Drive

print("Extraction terminée !")
```

```python
# On entraine le modele sur le jeu Train avec les parametres que j'ai defini
# dans le but d'ameliorer à la fois la performence et la duree du modele

results = model.train(data=dataset_path + "/data.yaml", epochs=100 , imgsz=800 , batch=16 , workers=4)
```

```python
# Tester notre modele sur les données de test définies dans data.yaml
import glob # Pour acceder aux images test sans pour autant changer mon fichier YAML parceque model.test() n'exite pas

# Charger le modèle
model = YOLO("runs/detect/train/weights/best.pt") # c'est le chemin de notre model Yolov8 deja entraine

# Récupérer les images de test
test_images = glob.glob("/content/drive/MyDrive/archive/test/images/*.jpg")  # Adapte selon notre jeu deonnées test
print("Nombre d'images trouvées :", len(test_images))
print("Liste des images :", test_images)

# Faire des prédictions sur les données test
results = model(test_images)

# Afficher les résultats
for r in results:
    r.show()  # Afficher les images avec détection
```

```python
# Évaluer le modèle sur le jeu de validation
metrics = model.val(data=dataset_path + "/data.yaml")

# Afficher les résultats
print(metrics)
```
